```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
production = pd.read_csv('../data/cleaned_data/clean_production.csv')

production.head()
```

```python
# adding a column for Total production    
production['Total_production'] = production.iloc[:, 2:33].sum(axis=1)
```

```python
total_production = production[['Country', 'Total_production', 'Coffee type']].sort_values(by=['Total_production'],
                                                                                        ascending=False)
total_production.head(20)
```

```python
#looking for Top 5 producers
top5_production = production[['Country', 'Total_production', 'Coffee type']].sort_values(by=['Total_production'],
                                                                                        ascending=False).head(5)
top5_production
```

```python
top5_production.dtypes
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.barplot(x="Country",
           y=top5_production["Total_production"]/1000000 ,
           data=top5_production,
           errorbar=None,
#           hue="Coffee type"
           )
ax.set_ylabel("Total Coffee Production in kg (Billions)")
ax.set_xlabel("Coffee Producing Countries")
ax.set_title("Coffee Production Top 5 Producers (1990-2020)");

fig.savefig("../images/top5_producers.png")
```

```python
#Create df of production with only top 5 countries
production_time = production[production['Country'].isin(top5_production['Country'])]
#transpose df, making sure to sort values
production_time = production_time.sort_values(by=['Total_production'],ascending=False).T
#drop the non-numeric and totals rows
production_time.drop(['Country', 'Coffee type', 'Total_production'], axis=0, inplace=True)
# use top 5 country names
production_time.columns = top5_production['Country'].values
#set index name 
production_time.index.name = 'Year'
#reset the index
production_time.reset_index()
production_time
```

```python
production_time["Viet Nam"].sum()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
sns.lineplot(data=production_time,
            )
ax.set_title("Coffee Production Over Time for Top 5 Producers")
ax.set_ylabel("Coffee Production in kg (Millions)")
ax.xaxis.set_tick_params(rotation=90)
sns.despine();

#fig.savefig("../images/top5_over_time.png")
```
