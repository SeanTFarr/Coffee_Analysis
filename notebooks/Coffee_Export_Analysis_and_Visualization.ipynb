```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
export = pd.read_csv('../data/cleaned_data/clean_export.csv')

export.head()
```

```python
# adding a column for Total export    
export['Total_export'] = export.iloc[:, 2:31].sum(axis=1)
```

```python
total_export = export[['Country', 'Total_export']].sort_values(by=['Total_export'],
                                                                                        ascending=False)
total_export.head(20)
```

```python
total_export["Country"]
```

```python
#looking for Top 5 producers
top5_export = export[['Country', 'Total_export']].sort_values(by=['Total_export'],
                                                                                        ascending=False).head(5)
top5_export
```

```python
top5_export.dtypes
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.barplot(x="Country",
           y=top5_export["Total_export"]/1000000 ,
           data=top5_export,
           errorbar=None,
#           hue="Coffee type"
           )
ax.set_ylabel("Total Coffee Export in kg (Billions)")
ax.set_xlabel("Coffee Exporting Countries")
ax.set_title("Coffee Top 5 Exporters (1990-2020)")
ax.xaxis.set_tick_params(rotation=30);

fig.savefig("../images/top5_export.png")
```

```python
#Create df of export with only top 5 countries
export_time = export[export['Country'].isin(top5_export['Country'])]
#transpose df, making sure to sort values
export_time = export_time.sort_values(by=['Total_export'],ascending=False).T
#drop the non-numeric and totals rows
export_time.drop(['Country', 'Total_export'], axis=0, inplace=True)
# use top 5 country names
export_time.columns = top5_export['Country'].values
#set index name 
export_time.index.name = 'Year'
#reset the index
export_time.reset_index()
export_time
```

```python
export_time["Viet Nam"].sum()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
sns.lineplot(data=export_time,
            )
ax.set_title("Coffee Exporting Over Time for Top 5 Exporters")
ax.set_ylabel("Coffee Exports in kg (Millions)")
ax.xaxis.set_tick_params(rotation=90)
sns.despine();

fig.savefig("../images/top5_export_over_time.png")
```
