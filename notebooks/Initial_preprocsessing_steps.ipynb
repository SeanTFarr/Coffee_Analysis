### Production

```python
# the intial read in of production
production_df = pd.read_excel("../data/Total production.xlsx")

production_df.head()
```

```python
# I needed to place the header at the 3rd row
production_df = pd.read_excel("../data/Total production.xlsx", header=3)
# change 2 column names
production_df = production_df.rename(columns={"Unnamed: 1" : "Coffee type", "Crop year" : "Country"})
# drop NaNs
production_df.dropna(how='all', inplace=True)

production_df.head(10)
```

```python
#Looking at the top row, I need to see what other rows do not have a coffee type
production_df['Coffee type'].isna()
```

```python
# drop rows with NaN, then reset the index
production_df.drop(index=[1, 17, 26, 62, 63], axis=0, inplace=True)
production_df.reset_index(drop=True)
production_df.head()
```

```python
# replace the coffee types with something more readable
production_df.replace({"(A)" : "Arabica",
                       "(R)" : "Robusta",
                       "(R/A)" : "Robusta/Arabica",
                       "(A/R)" : "Arabica/Robusta"},
                      inplace=True)
print(production_df.shape)
production_df.head()
```

```python
production_df.info()
```

```python
#Rounding to nearest whole number
production_df = production_df.round(0)
#converting to integers
production_df = production_df. apply(lambda x: x. astype(int) if x. dtype == 'float' else x)
```

```python
production_df.head()
```

```python
production_df.info()
```

```python
# to get to showing numeric values by the thousand kg, multplying values by the 60 kg bag value
production_df = production_df. apply(lambda x: x * 60 if x. dtype == 'int' else x)

production_df.head()
```
