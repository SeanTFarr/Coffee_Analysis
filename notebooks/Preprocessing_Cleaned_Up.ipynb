```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore', category=UserWarning, module='openpyxl')
```

### Production

```python
# the read in of production
production_df = pd.read_excel("../data/Total production.xlsx")
# I needed to place the header at the 3rd row
production_df = pd.read_excel("../data/Total production.xlsx", header=3)
# change 2 column names
production_df = production_df.rename(columns={"Unnamed: 1" : "Coffee type", "Crop year" : "Country"})
# drop NaNs
production_df.dropna(how='all', inplace=True)
# drop rows with NaN, then reset the index
production_df.drop(index=[1, 17, 26, 62, 63], axis=0, inplace=True)
production_df.reset_index(drop=True)
# replace the coffee types with something more readable
production_df.replace({"(A)" : "Arabica",
                       "(R)" : "Robusta",
                       "(R/A)" : "Robusta/Arabica",
                       "(A/R)" : "Arabica/Robusta"},
                      inplace=True)
#Rounding to nearest whole number
production_df = production_df.round(0)
#converting to integers
production_df = production_df. apply(lambda x: x. astype(int) if x. dtype == 'float' else x)
# toget to showing numeric values by the thousand kg, multplying values by the 60 kg bag value
production_df = production_df. apply(lambda x: x * 60 if x. dtype == 'int' else x)

production_df.head()
```

### Imports

```python
# read in Imports
import_df = pd.read_excel("../data/Imports.xlsx", header=3)
# change column name
import_df = import_df.rename(columns={"Calendar years" : "Country"})
# drop NaNs
import_df.dropna(how='all', inplace=True)
# dropping the "Total" row and the copyright NaN row
import_df.drop(index=[39,40], axis=0, inplace=True)
import_df.reset_index(drop=True)
#rounding the data to whole number and filling NaNs with "0"
import_df = import_df.round(0)
import_df =import_df.fillna(0)
#converting floats to integers
import_df = import_df.apply(lambda x: x. astype(int) if x. dtype == 'float' else x)
# to get to showing numeric values by the thousand kg, multplying values by the 60 kg bag value
import_df = import_df. apply(lambda x: x * 60 if x. dtype == 'int' else x)

import_df.head()
```

### Exports

```python
#  read in of Exports, with change to header
export_df = pd.read_excel("../data/Exports - calendar year.xlsx", header=3)
# change column name
export_df = import_df.rename(columns={"Calendar years" : "Country"})
# drop NaNs
export_df.dropna(how='all', inplace=True)
#rounding the data to whole number and filling NaNs with "0"
export_df = export_df.round(0)
export_df =export_df.fillna(0)
#converting floats to integers
export_df = export_df.apply(lambda x: x. astype(int) if x. dtype == 'float' else x)
# to get to showing numeric values by the thousand kg, multplying values by the 60 kg bag value
export_df = export_df. apply(lambda x: x * 60 if x. dtype == 'int' else x)
export_df.head()
```

   ### Consumption    

```python
# read in of Consumption with header change
consumption_df = pd.read_excel("../data/Domestic consumption.xlsx", header=3)
# change 2 column names
consumption_df = consumption_df.rename(columns={"Unnamed: 1" : "Coffee type", "Crop year" : "Country"})
# drop NaNs
consumption_df.dropna(how='all', inplace=True)
# drop "Coffee type" rows with NaN, then reset the index
consumption_df.drop(index=[1, 17, 26, 62, 63], axis=0, inplace=True)
consumption_df.reset_index(drop=True)
# replace the coffee types with something more readable
consumption_df.replace({"(A)" : "Arabica",
                       "(R)" : "Robusta",
                       "(R/A)" : "Robusta/Arabica",
                       "(A/R)" : "Arabica/Robusta"},
                      inplace=True)
#rounding the data to whole number and filling NaNs with "0"
consumption_df = consumption_df.round(0)
consumption_df =consumption_df.fillna(0)
#converting floats to integers
consumption_df = consumption_df.apply(lambda x: x. astype(int) if x. dtype == 'float' else x)
# to get to showing numeric values by the thousand kg, multplying values by the 60 kg bag value
consumption_df = consumption_df. apply(lambda x: x * 60 if x. dtype == 'int' else x)
consumption_df.head()
```

### RE-exports

```python
# read in of Re-exports, with header change
reexport_df = pd.read_excel("../data/Re-exports.xlsx", header=3)
# change column name
reexport_df = import_df.rename(columns={"Calendar years" : "Country"})
# drop NaNs
reexport_df.dropna(how='all', inplace=True)
#rounding the data to whole number and filling NaNs with "0"
reexport_df = reexport_df.round(0)
reexport_df =reexport_df.fillna(0)
#converting floats to integers
reexport_df = reexport_df.apply(lambda x: x. astype(int) if x. dtype == 'float' else x)
# to get to showing numeric values by the thousand kg, multplying values by the 60 kg bag value
reexport_df = reexport_df. apply(lambda x: x * 60 if x. dtype == 'int' else x)

reexport_df.head()
```
